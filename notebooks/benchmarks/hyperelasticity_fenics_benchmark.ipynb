```python
from pathlib import Path
import json
import subprocess

REPO_ROOT = Path.cwd()
ARTIFACT_ROOT = Path('artifacts/raw_results/notebook_runs')
(REPO_ROOT / ARTIFACT_ROOT).mkdir(parents=True, exist_ok=True)
PYTHON = './.venv/bin/python'

def dump_json(path):
    with open(REPO_ROOT / path, encoding='utf-8') as handle:
        return json.load(handle)
```

```python
from pprint import pprint

out_dir = ARTIFACT_ROOT / 'hyperelasticity_fenics_benchmark'
(REPO_ROOT / out_dir).mkdir(parents=True, exist_ok=True)

def run_case(name, argv):
    out_path = out_dir / f'{name}.json'
    cmd = argv + [str(out_path)]
    print('$', ' '.join(cmd))
    subprocess.run(cmd, cwd=REPO_ROOT, check=True)
    payload = dump_json(out_path)
    return {
        'case': name,
        'result': payload.get('result'),
        'steps': len(payload.get('steps', [])),
        'total_newton_iters': payload.get('total_newton_iters'),
        'time': payload.get('time'),
        'final_energy': payload.get('steps', [{}])[-1].get('energy') if payload.get('steps') else None,
    }

results = [
    run_case('custom_serial', [PYTHON, 'src/problems/hyperelasticity/fenics/solve_HE_custom_jaxversion.py', '--level', '1', '--steps', '4', '--total-steps', '24', '--quiet', '--out']),
    run_case('snes_serial', [PYTHON, 'src/problems/hyperelasticity/fenics/solve_HE_snes_newton.py', '--level', '1', '--steps', '4', '--total_steps', '24', '--quiet', '--out']),
]
pprint(results)
```
